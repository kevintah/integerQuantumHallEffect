Gauges and gauge fixing

Starting from the relationship between the magnetic field and the vector potential and moving from there

$B = \nabla \times A$

Taking determinants


$\frac{\partial A_z}{\partial y} - \frac{\partial A_y}{\partial z} =  B_x$

$\frac{\partial A_x}{\partial z} - \frac{\partial A_y}{\partial x} =  B_y$

$\frac{\partial A_y}{\partial x} - \frac{\partial A_x}{\partial y} =  B_z$

It is left to solve the equations by grouping and adding. 



The hamiltonian for a charged particle in a magnetic field is 

$H = (\frac{ \vec{p} - q\vec{A}}{2m})^2 $

We are essentially going to be in 2d so

One of the possible solutions is the Landau gauge.

Landau Gauge :  $(-By,0,0)$

$H = (\frac{1}{2m} (p_x \hat{x} + p_y{\hat{y}}) -qBy \hat{x})^2 $

$H = (\frac{1}{2m} (p_x -qBy)^2 + p_y^2) $

move to operator for quantum

$H = (\frac{1}{2m} (P_x -qBy)^2 + P_y^2) $

$H \psi = E \psi$ 

Guess: $\psi(x,y) ~ \psi(y)e^{ik_xx}$

$P_x = \hbar k_x$

$H = (\frac{1}{2m} (\hbar k_x -qBy)^2 + P_y^2) $

$H = \frac{1}{2m}P_y^2 + \frac{m}{2}\frac{qB}{m^2}(y - (- \frac{\hbar k_x}{qB}))^2$

Let $y_0 =- \frac{\hbar k_x}{qB}$

$H = \frac{1}{2m}P_y^2 + \frac{m}{2}((\frac{qB}{m})^2(y - y_0)^2$

cyclotron frequency:

$\frac{m v^2}{r} = qBv$

$\omega = \frac{v}{r} = \frac{qB}{m}$

so

$H = \frac{1}{2m}P_y^2 + \frac{m}{2}\omega^2(y - y_0)^2$


Generally the characteristic length:

$l = \sqrt{\frac{\hbar}{m \omega}}$

$l_B = \sqrt{\frac{\hbar}{ qB}}$

$l_B$ here is now called the magnetic length.



Comparing $l_B$ and $y_0$:

$y_0 = -k_x l_B^2$


If all this is true , then this is a type of harmonic oscillator and it is possible to write down the energies and wave function.


$E = (\hbar \omega (n + \frac{1}{2})$

$\vec{I} = -q \vec{\dot{x}} $

$m \vec{\dot{x}} = \vec{p} + q \vec{A}$

$\vec{I}= \frac{-q}{m} \sum \langle \psi| -i \hbar \nabla + q \vec{A}|\psi \rangle$

It seems you can get edge states if you add a confining potential to the hamiltonian. You get modified solutions.
